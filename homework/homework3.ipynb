```python
import numpy as np
import matplotlib.pyplot as plt
#setting the range
n_values = np.arange(1, 10001)
# Calculating probability that jth observation is in the bootstrap sample for each n
probability_in_sample = 1 - (1 - 1/n_values)**n_values

# Creating the plot
plt.figure(figsize=(10, 6))
plt.plot(n_values, probability_in_sample, color='cyan')
plt.title('Probability of the $j$th Observation Being in the Bootstrap Sample', color='white')
plt.xlabel('Sample Size ($n$)', color='white')
plt.ylabel('Probability', color='white')
plt.xscale('log')  
plt.grid(True, which="both", ls="--", color='gray')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create an array of probability values from 0 to 1
p_mk = np.linspace(0.01, 0.99, 500)  # Avoiding 0 and 1 for entropy calculation
# Calculate p_m1 and p_m2
p_m1 = p_mk
p_m2 = 1 - p_mk

# Calculate the Gini index
gini_index = 2 * p_m1 * p_m2
# Calculate the classification error
classification_error = 1 - np.maximum(p_mk, p_m2)
# Calculate the entropy
entropy = -p_m1 * np.log2(p_m1) - p_m2 * np.log2(p_m2)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(p_mk, gini_index, label='Gini Index', color='cyan')
plt.plot(p_mk, classification_error, label='Classification Error', color='magenta')
plt.plot(p_mk, entropy, label='Entropy', color='yellow')

# Plot formatting
plt.title('Gini Index, Classification Error, and Entropy as functions of $\hat{p}_{mk}$', color='white')
plt.xlabel('$\hat{p}_{mk}$', color='white')
plt.ylabel('Value', color='white')
plt.legend()
plt.grid(True, which="both", ls="--", color='gray')
plt.show()
```
